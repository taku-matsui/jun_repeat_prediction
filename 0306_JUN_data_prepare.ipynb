```python
# coding: UTF-8
import pandas as pd
import numpy as np
import datetime
# 読み込み
print("loading...")
df_jun = pd.read_csv('jun_needed_data.csv').iloc[:,1:] 
# 欠損値削除と無効なmember_idを削除
print("Removing invalid data...")
df_jun_valid = df_jun.dropna().query('member_id != 0')
```

```python
# 無効な誕生年を削除（1900年～2020年のみ抽出）
print("Removing unappropriate birthday...")
mask = (df_jun_valid.birthday.astype(np.int) > 19000000) & (df_jun_valid.birthday.astype(np.int) < 20200000)
df_jun_final = df_jun_valid[mask]
# 誕生日と購入日を文字列に変換
print("Converting the birthday and sales date to 'str'...")
df_jun_final['birthday'] = df_jun_final['birthday'].astype(str)
df_jun_final['sales_date'] = df_jun_final['sales_date'].astype(str)
```

```python
# birthdayのうち無効な日付を削除
print("Removing invalid date from 'birthday'...")
#year = df_jun_final.birthday.str.slice(0,4).astype(int)
month = df_jun_final.birthday.str.slice(4,6).astype(int)
day = df_jun_final.birthday.str.slice(6,8).astype(int)
df_jun_final = df_jun_final[(month > 0) & (month < 13) & (day > 0) & (day < 32)]
print("Number of the valid data is...")
print(len(df_jun_final))
```

```python
# 購入日をDATETIMEに変換
print("Converting sales data to 'datetime'...")
today = pd.DataFrame(df_jun_final['sales_date'].astype(float).astype(np.int))
today.sales_date = pd.to_datetime(today.sales_date, format='%Y%m%d')
df_jun_final['sales_date'] = today
# 誕生日をDATETIMEに変換
print("Converting birthday to 'datetime'...")
birth = pd.DataFrame(df_jun_final['birthday'].astype(float).astype(np.int))
birth.birthday = pd.to_datetime(birth.birthday, format='%Y%m%d')
df_jun_final['birthday'] = birth
```

```python
window = 3 # 一つの入力に用いる日数
table = []
train_data = []
for cid, tbl in df.tail(20).groupby('member_id'):
    # あるユーザ（顧客ID: cid）に対して1回のループ
    buy_date = tbl.sales_date.values
    # その日は経過日数を"-1"とする
    d = [-1] + [(buy_date[i] - buy_date[i - 1]).days for i in range(1, len(buy_date))]
    tbl_tmp = tbl.assign(days_delta=d) # days_delta（前の購入日との差）カラムを追加
    
    # 以下、d_tmpに対して
    # cid, shop_id, age, days_delta, gender. pref_cd, come_flag
    for i in range(tbl_tmp.shape[0] - 1): # (ある顧客のデータ数 - 1)だけ
        # ある顧客のデータに対して1回のループ
        # "i" が注目している購入日を表す
        # 購入日(sales_date)は古い順に並んでいる
        d_tmp = [cid] # まずは顧客ID
        for w in range(window): # i > window になったら穴埋め
            if i - w >= 0:
                d_tmp += [tbl_tmp.iloc[i - w].age,
                          tbl_tmp.iloc[i - w].days_delta
                         ]
            else:
                d_tmp += ['-', '-']
        d_tmp.append(tbl_tmp.iloc[0].gender) # 年齢
        d_tmp.append(tbl_tmp.iloc[0].pref_cd) # 住所
        
        # 以降の購入が2週以降6週以内  --> True
        days = tbl_tmp.iloc[i:].days_delta.values
        days = [np.where(days > 0)] # days < 0 のインデックスを取得
        ds = (14 <= np.cumsum(days)) * (np.cumsum(days) <= 42)
        d_tmp.append(1 if True in ds else 0) # 購入があれば 1, なければ 0
        train_data.append(d_tmp)
        
    print("customer %s" %cid)
    print(tbl_tmp)
```
